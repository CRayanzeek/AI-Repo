```python
#importing embedding model
embedding_model = SentenceTransformer(model_name_or_path="all-mpnet-base-v2")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
embedding_model.to(device)
```

```python
# Cell 2: Functions Definitions
def create_title_embedding(title):
    title_embedding = embedding_model.encode(title, convert_to_tensor=True, device=device)
    return title_embedding

def load_detailed_embeddings(file_path):
    with open(file_path, 'r') as f:
        loaded_data = json.load(f)
    tensor_embeddings_list = loaded_data["embeddings"]
    pages_and_chunks = loaded_data["pages_and_chunks"]
    
    tensor_embeddings = torch.tensor(tensor_embeddings_list).to(device)
    
    return tensor_embeddings, pages_and_chunks

def create_book_embedding(embeddings):
    book_embedding = torch.mean(embeddings, dim=0)
    return book_embedding

def create_combined_book_embedding(title, embeddings):
    title_embedding = create_title_embedding(title)
    content_embedding = create_book_embedding(embeddings)
    combined_embedding = torch.mean(torch.stack([title_embedding, content_embedding]), dim=0)
    return combined_embedding

def load_books(subject_books):
    book_embeddings = []
    book_metadata = []
    
    for book_path in subject_books:
        with open(book_path, 'r') as f:
            book_data = json.load(f)
            title = book_data['book_title']
        
        embeddings, pages_and_chunks = load_detailed_embeddings(book_path)
        combined_embedding = create_combined_book_embedding(title, embeddings)
        
        book_embeddings.append(combined_embedding)
        book_metadata.append({'title': title, 'file_path': book_path, 'pages_and_chunks': pages_and_chunks})
    
    return torch.stack(book_embeddings), book_metadata
```

```python

all_subject_embeddings = {}
all_subject_metadata = {}

for subject, books in Subjects.items():
    embeddings, metadata = load_books(books)
    all_subject_embeddings[subject] = embeddings
    all_subject_metadata[subject] = metadata
```

```python
# Cell 4: Create Subject-Level Embeddings
def create_subject_embeddings(subject_embeddings):
    subject_means = {}
    for subject, embeddings in subject_embeddings.items():
        subject_means[subject] = torch.mean(embeddings, dim=0)
    return subject_means

subject_means = create_subject_embeddings(all_subject_embeddings)
```

```python
tokenizer = tiktoken.encoding_for_model("gpt-3.5-turbo")
```

```python

def count_tokens(text):
    return len(tokenizer.encode(text))
```

```python
def retrieve_relevant_resources(query, embeddings, model=embedding_model, n_resources_to_return=5):
    query_embedding = model.encode(query, convert_to_tensor=True, device=device)
    
    # Ensure embeddings are on the same device as the query embedding
    if embeddings.device != query_embedding.device:
        embeddings = embeddings.to(query_embedding.device)
    
    dot_scores_final = util.dot_score(query_embedding, embeddings)[0]
    
    n_resources_to_return = min(n_resources_to_return, len(dot_scores_final))
    scores, indices = torch.topk(dot_scores_final, k=n_resources_to_return)

    return scores, indices

def chunk_context(context_items, max_length=4096):
    chunks = []
    current_chunk = ""
    current_chunk_tokens = 0

    for item in context_items:
        item_text = item if isinstance(item, str) else item["sentence_chunk"]
        item_tokens = count_tokens(item_text)

        if current_chunk_tokens + item_tokens + 2 > max_length:
            chunks.append(current_chunk.strip())
            current_chunk = ""
            current_chunk_tokens = 0

        current_chunk += "- " + item_text + "\n"
        current_chunk_tokens += item_tokens + 2

    if current_chunk:
        chunks.append(current_chunk.strip())

    return chunks

def prompt_formatter(query, context_items, max_length=4096):
    context_chunks = chunk_context(context_items, max_length)

    prompts = []
    for chunk in context_chunks:
        prompt = f"""Based on the following context items, please answer the query
        Context items:
        {chunk}
        Query: {query}
        Answer:
        """
        prompts.append(prompt)
    return prompts

def search_books(query):
    # Step 1: Find the most relevant subject
    subject_embeddings = torch.stack(list(subject_means.values()))
    subject_embeddings = subject_embeddings.to(device)
    subject_scores, subject_indices = retrieve_relevant_resources(query, subject_embeddings)
    
    # Step 2: Within the most relevant subject, find the most relevant books
    relevant_subject = list(subject_means.keys())[subject_indices[0]]
    book_embeddings = torch.stack([torch.tensor(all_subject_embeddings[relevant_subject][i]) for i in range(len(all_subject_embeddings[relevant_subject]))])
    book_embeddings = book_embeddings.to(device)
    book_scores, book_indices = retrieve_relevant_resources(query, book_embeddings)
    
    # Step 3: Detailed search within the most relevant books
    final_context_items = []
    for book_idx in book_indices:
        book_metadata = all_subject_metadata[relevant_subject][book_idx]
        book_embeddings, pages_and_chunks = load_detailed_embeddings(book_metadata['file_path'])
        book_embeddings = book_embeddings.to(device)
        page_scores, page_indices = retrieve_relevant_resources(query, book_embeddings)
        
        for page_idx in page_indices:
            final_context_items.append(pages_and_chunks[page_idx])
    
    prompts = prompt_formatter(query, final_context_items)
    return prompts





```

```python

def rank_context_items(query, context_items, model=embedding_model):
    query_embedding = model.encode(query, convert_to_tensor=True, device=device)
    context_embeddings = model.encode([item if isinstance(item, str) else item["sentence_chunk"] for item in context_items], convert_to_tensor=True, device=device)
    
    scores = util.dot_score(query_embedding, context_embeddings)[0]
    ranked_indices = torch.argsort(scores, descending=True)

    ranked_context_items = [context_items[i] for i in ranked_indices]
    return ranked_context_items
```

```python
def generate_response(prompt: str, api_key: str, model: str = "gpt-3.5-turbo",temperature: float = 0.1):
    url = "https://api.openai.com/v1/chat/completions"
    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json"
    }
    body = {
        "model": model,
        "messages": [
            {"role": "system", "content": "You are a helpful assistant knowledgeable about finance, who is goes into detail about subject."},
            {"role": "user", "content": prompt}
        ],
        "temperature": temperature
    }
    body_json = json.dumps(body)
    
    max_retries = 5
    retry_delay = 1

    for attempt in range(max_retries):
        response = requests.post(url, headers=headers, data=body_json)
        if response.status_code == 200:
            response_json = response.json()
            return response_json["choices"][0]["message"]["content"]
        elif response.status_code == 429:
            print(f"Rate limit exceeded. Retrying in {retry_delay} seconds...")
            time.sleep(retry_delay)
            retry_delay *= 2
        else:
            print(f"Error: {response.status_code}")
            break
    else:
        print("Failed after several retries.")
        return None

# Main logic
query = "explain entrepreneurial finance"
context_items = search_books(query)

# Rank context items
ranked_context_items = rank_context_items(query, context_items)

# Print ranked context items
print("Ranked Context Items:")
for idx, item in enumerate(ranked_context_items):
    print(f"{idx + 1}: {item if isinstance(item, str) else item['sentence_chunk']}")

# Select top context item
top_context_item = ranked_context_items[0] if isinstance(ranked_context_items[0], str) else ranked_context_items[0]["sentence_chunk"]

# Print the selected top context item
print("\nSelected Top Context Item:")
print(top_context_item)

# Format the prompt
prompt = prompt_formatter(query, [top_context_item])[0]

# Print the formatted prompt to confirm its content
print("\nFormatted Prompt:")
print(prompt)


# Generate response from GPT-3.5 Turbo
response = generate_response(prompt, api_key)
print(f"\nResponse:\n{response}")
```
